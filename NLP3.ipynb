```python
def minimum_edit_distance(s1, s2):
    m = len(s1)
    n = len(s2)

    # Create DP table
    dp = [[0 for _ in range(n + 1)] for _ in range(m + 1)]

    # Base cases
    for i in range(m + 1):
        dp[i][0] = i

    for j in range(n + 1):
        dp[0][j] = j

    # Fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):

            # If characters are same
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]

            # If characters are different
            else:
                dp[i][j] = 1 + min(
                    dp[i][j - 1],     # Insert
                    dp[i - 1][j],     # Delete
                    dp[i - 1][j - 1]  # Replace
                )

    # Print DP Table (Optional)
    print("DP Table:")
    for row in dp:
        print(row)

    return dp[m][n]


# Driver Code
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

result = minimum_edit_distance(s1, s2)

print("Minimum Edit Distance =", result)
```
